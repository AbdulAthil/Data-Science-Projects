# Data Visualization

### Plot the number of students for each school and categorize it by sex: (Hint: countplot)

```python
sns.countplot(x='school',data=data_student,hue="sex")
```

### Plot age distribution of the students(include kde)

```python
sns.distplot(x=data_student['age'],kde=True)
```

### Plot average of students for each school:

```python
sns.barplot(x='school',y='age',data=data_student)
```

### Plot the quartiles of final grade(G3) of the students categorized by mother's education(Medu) (Hint:boxplot)

info:

0 - None

1 - primary education(4th grade)

2 - 5th to 9th grade

3 - secondary education

4 - higher education

```python
sns.boxplot(x="Medu",y="G3",data=data_student)
```

### Plot probability density with regards to final grade (G3) and mother's job (Mjob) and consider sex by selecting split=True:  (Hint:violinplot)

```python
#using violinplot  -- we use violinplot because only violin plot has split attribute
sns.violinplot(x="Mjob",y="G3",data=data_student,hue="sex",split=True)
```

```python
#using barplot
sns.barplot(x="Mjob",y="G3",data=data_student,hue="sex")
```

```python
#using boxplot
sns.boxplot(x="Mjob",y="G3",data=data_student,hue="sex")
```

### Plot density distribution of the grades (both G2,G3 on the same canvas) with regards to health but this time we want to see every data as a point in our plot: (Hint: swarmplot,subplot)

info: health status(numeric: from 1 - very bad to 5 - very good)

```python
fig,axes=plt.subplots(1,2,figsize=(13,5))

sns.swarmplot(x="health",y="G2",data=data_student,ax=axes[0])
sns.swarmplot(x="health",y="G3",data=data_student,ax=axes[1])
```

### Do the same for the absences and travel time (single plot is enough)

```python
 sns.swarmplot(x="traveltime",y="absences",data=data_student)
```

### Plot the final grade (G3) with respect to the parental status(Pstatus) and family size(famsize) using FacetGrid and distplot (set kde=False)

info:


LE3 - less than and equal to 3

GT3 - Greater than 3

A - Apart

T - Together

```python
facet= sns.FacetGrid(data_student,row="Pstatus",col="famsize")

facet.map(sns.distplot,"G3",kde=False)
```

### Plot a categorical relationship between study time(studytime) and mean of final grade (G3) (Hint: barplot)

```python
sns.barplot(x="studytime",y="G3",data=data_student,estimator=np.mean)
```

### Plot a relationship between max, midterm grade(G2) and reason for closing the school(reason) and categorize it by sex:        (Hint: We need barplot but let us use catplot to reach the same plot)

```python
#using barplot
sns.barplot(x="reason",y="G2",data=data_student,hue="sex",estimator=np.max)
```

```python
#using catplot
sns.catplot(x="reason",y="G2",data=data_student,kind="bar",hue="sex",estimator=np.max)
```

### Plot the distribution of the final grade(G3) of the students using kde. Plot for both gender type on the same plot:  (Hint: distplot using logical masking could come in handy.)

```python
sns.distplot(data_student[data_student['sex']=="M"]["G3"],label='Male',color="red",hist=True)
sns.distplot(data_student[data_student['sex']=="F"]["G3"],label="Female",color="yellow",hist=True)
```

### This time plot the distribution of midterm grade(G2) using kde together for each freetime amount: (Hint: Previously we had 2 plot together, this time we had 5, so let's use for loop instead)

```python
data_student["freetime"].nunique()
```

```python
for time in data_student["freetime"].unique():
    sns.distplot(data_student[data_student["freetime"]==time]["G3"],label=time,hist=False)
```

### Create a jointplot for relationship between midterm(G2) and final grades(G3):

```python
sns.jointplot(x="G2",data=data_student,y="G3",kind="kde")
```

### Let's do the same for absences and midterm grade (G2) but this time with hex:

```python
sns.jointplot(x="absences",y="G2",data=data_student,kind="hex")
```

### Plot the relationship between midterm(G2),final grade(G3) and absences:  (Hint: we have more than 2 columns to pair)

```python
sns.pairplot(data_student[['G2','G3','absences']])
```
