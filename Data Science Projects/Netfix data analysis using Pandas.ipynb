Unfortunately,there are no IMDB ratings in the dataset

We will create a random array from 0 to 10 which has the same size to insert into the dataframe as 'IMDB' ratings.
        Important: Because we are using a random array, there is actually no correct result to the tasks involving IMDB ratings.
        Everyone will see a different ratings

Feel free to run the line below to generate the IMDB ratings

```python
#Insert the IMDB rating column in columns
```

```python
IMDB = np.random.uniform(low=0,high=10, size=len(df["title"]))
IMDB = np.around(IMDB,decimals=1)
```

Insert the random ratings we created as a new column into the DataFrame and name it as 'IMDB' ratings

```python
df['IMDB rating']=IMDB
df.columns
#Now the IMDB rating column added with the existing columns
```
